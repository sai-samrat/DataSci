```python
t=np.arange(0.0,24,0.01)
s=np.sin(t)
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='radians',ylabel='Sin(x)',title='Sine plot')
ax.grid()
plt.show()
```

```python
s=np.cos(t)
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='radians',ylabel='Sin(x)',title='Cosine plot')
ax.grid()
plt.show()
```

```python
s=np.tan(t)
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='radians',ylabel='Sin(x)',title='Tangent plot')
ax.grid()
plt.show()
```
